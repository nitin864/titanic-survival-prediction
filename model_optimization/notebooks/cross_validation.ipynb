```python
# imports
import pandas as pd
import numpy as np
```

```python
# sklearn imports
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import joblib
import json
```

```python
# loading data
df = pd.read_csv('../../cleaned_data/titanic_featured.csv')
print(df.shape)
```

```python
# features and target
X = df.drop('Survived', axis=1)
y = df['Survived']
```

```python
# model
model = LogisticRegression(max_iter=200)
```

```python
# cross validation
scores = cross_val_score(model, X, y, cv=5)
print(scores)
```

```python
# average score
avg_score = scores.mean()
print('Average Score:', round(avg_score, 4))
```

```python
# saving results
cv_results = {
    'scores': scores.tolist(),
    'average_score': round(avg_score, 4)
}

with open('../results/cross_validation_results.json', 'w') as f:
    json.dump(cv_results, f, indent=2)

joblib.dump(model, '../results/cross_validation_model.pkl')

print('Results saved')
```
