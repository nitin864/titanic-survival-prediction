```python
# imports
import pandas as pd
import numpy as np
```

```python
# loading data
df = pd.read_csv('../../cleaned_data/titanic_featured.csv')
```

```python
# creating some extra features

df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

df['FarePerPerson'] = df['Fare'] / df['FamilySize']

df['AgeFareRatio'] = df['Age'] / (df['Fare'] + 1)

df['IsChild'] = np.where(df['Age'] < 16, 1, 0)

df['IsRich'] = np.where(df['Fare'] > df['Fare'].median(), 1, 0)
```

```python
# saving feature engineered dataset
df.to_csv('../results/feature_engineered.csv', index=False)
print('Feature engineered dataset saved')
```
