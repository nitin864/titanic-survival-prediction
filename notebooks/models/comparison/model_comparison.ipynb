```python
#importing libraries
import json
import pandas as pd
import matplotlib.pyplot as plt
```

loading results

```python
#loading both model results
with open('../results/logistic_regression.json') as f:
    lr = json.load(f)['logistic_regression']

with open('../results/random_forest.json') as f:
    rf = json.load(f)['random_forest']
```

METRICS TABLE

```python
#comparison table for both_models
df = pd.DataFrame({'logistic_regression': lr, 'random_forest': rf}).T
df
```

ACCURACY COMPARISON

```python
#accuracy bar chart
models = ['logistic_regression', 'random_forest']
accs   = [lr['accuracy'], rf['accuracy']]

plt.figure(figsize=(5,4))
bars = plt.bar(models, accs, color=['#3498db', '#2ecc71'], edgecolor='black')
plt.ylim([0.75, 0.85])
plt.title('accuracy_comparison')
plt.ylabel('accuracy')

for bar, val in zip(bars, accs):
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 0.002,
             f'{val:.4f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

ROC-AUC COMPARISON

```python
#roc_auc bar chart
aucs = [lr['roc_auc'], rf['roc_auc']]

plt.figure(figsize=(5,4))
bars = plt.bar(models, aucs, color=['#3498db', '#2ecc71'], edgecolor='black')
plt.ylim([0.82, 0.87])
plt.title('roc_auc_comparison')
plt.ylabel('roc_auc')

for bar, val in zip(bars, aucs):
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 0.001,
             f'{val:.4f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
#winner by accuracy
if lr['accuracy'] > rf['accuracy']:
    print(f'winner: logistic_regression  accuracy: {lr["accuracy"]}')
else:
    print(f'winner: random_forest  accuracy: {rf["accuracy"]}')

print(f'accuracy difference: {abs(lr["accuracy"] - rf["accuracy"]):.4f}')
print(f'roc_auc difference : {abs(lr["roc_auc"]  - rf["roc_auc"]):.4f}')
```
